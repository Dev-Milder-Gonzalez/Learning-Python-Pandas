# ExploraciÃ³n inicial de datos con ```head()``` y ```tail()```ðŸ“˜

## 1. Firma de los mÃ©todosâœï¸
Ambos mÃ©todos esta disponibles para ```Series``` y ```DataFrames```. Su funciÃ³n principal es __retornar__ las primeras o Ãºltimas ```n``` filas.
### 1.1. Retornar las primeras ```n``` filasðŸ§±
```
DataFrame.head(n: int = 5) -> DataFrame
Serie.head(n: int = 5) -> Serie
```
### 1.1. Retornar las Ãºltimas ```n``` filasðŸ§©
Este mÃ©todo es Ãºtil para ver si se cargaron completamente o si hya filas de resumen/basura al final de archivo.
```
DataFrame.tail(n: int = 5) -> DataFrame
Series.tail(n: int = 5) -> Series
```
## 2. ParÃ¡metros
 - __Por defecto (``` n = a```):__ En el caso de que no se especifique nada. Pandas muestra 5 filas.
 - __Entero positivo(```n = a```):__ Muestra exactamente esa cantidad de filas, segÃºn el mÃ©todo.
 - __Entero negativo():__ 
  - __head(-a):__ Muestra todas las filas excepto las Ãºltimos ```n```. El __a__ es un entero positivo.
  -__tail(-a):__ Muestra todas las filas excepto las primera ```n```.Se cumple ```a  >= 0```



````python
import pandas as pd
import numpy as np
````

````python
df_iris = pd.read_csv("../../../data/raw/iris_DataSet.csv")
print(f'Dimesiones del DataFrame:\n{df_iris.shape}')
print('=' * 80)
print(df_iris)
````

## 3. MÃ©todo: ```head()```

````python
# Primeras 5 filas
p_cinco_filas = df_iris.head()
print(f'Dimensiones:\n{p_cinco_filas.shape}')
print('=' * 60)
print(f'Primera cinco filas (0 - 4):\n{p_cinco_filas}')
````

````python
# Primero 30 filas
p_treinta_filas = df_iris.head(20)
print(f'Dimensiones:\n{p_treinta_filas.shape}')
print('=' * 60)
print(f'Primeras veinte filas (0 -19):\n{p_treinta_filas}')
````

````python
# Todas las filas menos las "n" Ãºltimas
df_head_negativo = df_iris.head(-20)
print(f'Dimesiones:\n{df_head_negativo.shape}')
print('=' * 70)
print(f'DataFrame:\n{df_head_negativo}')
````

## 3. MÃ©todo: ```tail()```

````python
# Las Ãºltimas 5 filas
ultimas_cinco_filas = df_iris.tail()
print(f'Dimesiones:\n{ultimas_cinco_filas.shape}')
print('=' * 70)
print(f'Ãšltimas cinco filas:\n{ultimas_cinco_filas}')
````

````python
# Las Ãºltimas "n" filas
ultimas_veinte_filas = df_iris.tail(20)
print(f'Dimensiones:\n{ultimas_veinte_filas.shape}')
print('=' * 70)
print(f'Ãšltimas veinte filas:\n{ultimas_veinte_filas}')
````

````python
# Todas las filas menos las "n" primeras
# Las Ãºltimas "n" filas
df_tail_negativo = df_iris.tail(20)
print(f'Dimensiones:\n{df_tail_negativo.shape}')
print('=' * 70)
print(f'Ãšltimas veinte filas:\n{df_tail_negativo}')
````
