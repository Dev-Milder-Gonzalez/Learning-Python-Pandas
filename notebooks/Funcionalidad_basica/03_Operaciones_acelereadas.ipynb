# Operaciones aceleradas
Pandas incluye soporte para acelerar ciertos tipos de operaciones num茅ricas (binarias y  booleanas) utilizando las librer铆as ```numexpr``` y ```bottleneck```.
 - __```numexpr```:__ Usada para acelerar operaciones __binarias y booleanas__ especialmente para grandes conjutos de datos.
  - __```bottleneck```:__ Usado para acelerar ciertas __operaciones de reducci贸n__ (por ejemplo ```min```, ```max```, ```sum```, ```median```, entre otras).
Ambas librer铆as estas __habilitadas por defecto__. Podemos controlar su pandas las usa o no estableciendo las siguientes opciones globales.

````python
import pandas as pd
import numpy as np
````

````python
print('Desabilitando Aceleraci贸n:')
#Desabilitar el uso de bottleneck 
pd.set_option("compute.use_bottleneck", False)
#Desabilitar  el uso de numexpr
pd.set_option("compute.use_numexpr", False)
print(f'Bottleneck desabilitando:\n{pd.get_option("compute.use_bottleneck")}')
print(f'Numexpr desabilitando:\n{pd.get_option("compute.use_numexpr")}')
````

````python
print('Restaurando a Default (Habilitado):')
#Desabilitar el uso de bottleneck 
pd.set_option("compute.use_bottleneck", True)
#Desabilitar  el uso de numexpr
pd.set_option("compute.use_numexpr", True)
print(f'Bottleneck habilitando:\n{pd.get_option("compute.use_bottleneck")}')
print(f'Numexpr desabilitando:\n{pd.get_option("compute.use_numexpr")}')
````

## 1. Aceleraci贸n con ```numexpr``` (Operaciones bunarias y booleanas)
```numexpr``` acelera cualquer operacion que involucre dos o m谩s array de la misma forma (mismo objetos ```DataFrames``` o ```Series```).

````python
N = 10**6
df1 = pd.DataFrame(np.random.rand(N, 2))
df2 = pd.DataFrame(np.random.rand(N, 2))
print(f'Primer DataFrame:\n{df1.head(10)}')
print('=' * 35)
print(f'Segundo DataFrame:\n{df2.head(10)}')
````

````python
# Operaciones binaria
df_multiplicado = df1 * df2
print(f'Resultado:\n{df_multiplicado.head(15)}')
````

````python
# Operacion booleana
df_booleano = df1 > df2
print(f'Operacion booleana acelearada:\n{df_booleano.head(10)}')
````

````python
#Operaciones compleja
r_operacion_compleja = (df1 +df2) * (df1 * df2) + (df1 ** 2)
print(f'Operaci贸n compleja:\n{r_operacion_compleja.sample(20)}')
````

## 2. Aceleraci贸n con ```bottleneck``` (Operaci贸n de reducci贸n)З
Tambi茅n llamados m茅todos de resumen, dado que sintetizan un gran cantidad de datos en un solo valor. Especialmente cuando hay valores faltantes (```NaN```).

````python
N = 10 ** 6
serie_con_nulos = pd.Series(np.random.rand(N))
serie_con_nulos.iloc[::1000] = np.nan #Anadimos NaNs cada 100 filas
print(f'Serie:{serie_con_nulos.tail(10)}')
````

````python
#Reducci贸n con valores faltantes
suma = serie_con_nulos.sum()
media = serie_con_nulos.median()
minimo = serie_con_nulos.min()
conteo_nulos = serie_con_nulos.isna().sum()
print('Operaciones de reducci贸n aceleradas:')
print(f'Suma: {suma:.4f}')
print('=' * 35)
print(f'Media: {media:.4f}')
print('=' * 35)
print(f'Valo r m铆nimo: {minimo:.4f}')
print('=' * 35)
print(f'Cantidad de nulos:{conteo_nulos:.4f}')
````

__Resumen:__ No hay sintaxis especial de Pandas para "usar ```numexpr```". Simplemente escribimos el c贸digo est谩ndar (```df1 * df2```), y si la operaci贸n se puede optimizar, Pandas redirige la llamada al c贸digo acelerado de ```numexpr``` o ```bottleneck``` en lugar de usar las rutinas base de ```NumPy```.
