# Indexaci贸n y selecci贸n de filas b谩sicaЗ

````python
import pandas as pd
import numpy as np
````

````python
data = {
    'one': [1.0, 2.0, 3.0, 4.0],
    'bar': [1.0, 2.0, 3.0, 4.0],
    'flag': [False, False, True, False],
    'foo': ["bar", "bar", "bar", "bar"]
}

df = pd.DataFrame(data, index = ["fila_A", "fila_B", "fila_C", "fila_D"])
print(f'DataFrame base:\n{df}')
````

## 1. Seleccionar una columna por su etiqueta: ```df[etiqueta_columna]``` 1锔

````python
columna_one = df["one"]
print(f'Seleccionamos la columna "one":\n{columna_one}')
print('=' * 40)
print(f'Tipo:\n{ type(columna_one) }')
````

## 2. Selecci贸n de fila por etiqueta: ```df.loc[etiqueta_fila]```2锔

````python
fila_B = df.loc["fila_B"]
print(type(fila_B))
print('=' * 50)
print(f'Fila B:\n{fila_B}')
````

## 3. Selecci贸n de fila por su posici贸n entera: ```df.iloc[ubicaci贸n]```3锔

````python
fila_pos_0 = df.iloc[2]
print(type(fila_pos_0))
print('=' * 40)
print(f'Fila de la posici贸n 0:\n{fila_pos_0}')
````

## 4. Selecci贸n de filas por Slicing (rango entero de posiciones)З


````python
slice_filas = df[1:3]
print(type(slice_filas))
print('=' * 50)
print(f'Desde la primera fila hasta la tercera:\n{slice_filas}')
````

# Selecci贸n por vector booleano: ```df[bool_vector]```5锔
Es este modo de slecci贸n devuelve un Dataframe.

````python
df_seleccion_bool = df[ df["one"] * df["bar"] > 5]
print(f'DataFrame filtrado por condici贸n booleana:\n{df_seleccion_bool}')
````
