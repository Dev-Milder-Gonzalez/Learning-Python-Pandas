# Pantalla de la consola üñ•Ô∏è
Cuando se imprime un DataFrame en la consola, pandas los __trunca (corta)__ autom√°ticalemnete para que quepa en al pantalla y sea legible. Tambi√©n podemos obtener un resumen con ```info()```.

````python
import pandas as pd
import numpy as np
````

````python
data_sim = np.random.randint(0, 100, size=(100, 23))
column_names = [f'col_{i}' for i in range(23)]

df_largo = pd.DataFrame(data_sim, columns=column_names)

print("üìêDataFrame largo (pandas lo trunca autom√°ticamente):")
print(df_largo)
````

## 1. Resumen estructurado con ```info()``` üêº
Con m√©todo ```info()``` se realiza un resumen esructurado de los dos datos de un objeto de Pandas ya sea de una ```Serie``` un on ```DataFrame```.

````python
print("\nüí°Usando .info() para un resumen estructurado:")
df_largo.info()
````

# 2. Mostrar el dataFrame completo con el m√©todo:  ```df.to_string()```
Estos evita que el dataframe se trunque y se muestre completamente.

````python
print(df_largo.to_string())
````

## 3. Control de opciones de visualizaci√≥n

### 3.1. Ancho de visualizaci√≥n

````python
#‚öôÔ∏èReducimos el ancho de visualizaci√≥n (por defecto suele ser 80)
pd.set_option("display.width", 40) 
df_ancho = pd.DataFrame(np.random.randn(3, 10))

print("\n‚ÜîÔ∏èDataFrame impreso con ancho de consola reducido (40):")
print(df_ancho)
#Veremos que se necesitamos m√°s filas para mostrar las 10 columnas.
````

__üí°Consejo:__ Siempre restablecer a un valor alto para el trabajo normal.

````python
pd.set_option("display.width", 80)
````

### 3.2. Ancho m√°ximo de columna
Controlar cu√°ntos caracteres se muestran en una columna individual (limitar el n√∫mero de caracteres) antes de que el texto se trunque con ... Es crucial para columnas con strings largos (como paths o URLs).

````python
# üìù Datos con paths largos
datos_paths = {
    "nombre": ["file_01", "file_02"],
    "ruta_completa": [
        "media/user/almacenamiento/carpeta_larga_de_archivos_clave_01/file_01.txt",
        "media/user/almacenamiento/carpeta_larga_de_archivos_clave_02/file_02.txt",
    ],
}
df_paths = pd.DataFrame(datos_paths)
````

### 3.3. Truncamiento bajo (solo se ven los primeros n caracteres)üìè

````python
pd.set_option("display.max_colwidth", 30)
print("\nüìè Ancho de Columna = 30 (Truncado):")
print(df_paths)
````

 ### 3.4. Mostrar m√°s caracteres (n caracteres)

````python
pd.set_option("display.max_colwidth", 100)
print("\nüìè Ancho de Columna = 100 (Completo):")
print(df_paths)
````

### 3.4. Deshabilitar Salto de L√≠nea con la opci√≥n: ```display.expand_frame_repr``üö´
Por defecto, los DataFrames anchos se imprimen en varias filas. Al establecer esta opci√≥n en False, se fuerza a imprimir en un solo bloque horizontal, lo que puede requerir desplazamiento.

````python
# üö´ Deshabilitar el salto de l√≠nea (imprime el DataFrame en una sola "tira" horizontal)
pd.set_option("display.expand_frame_repr", False)
print("üö´ DataFrame ancho en una sola l√≠nea (requiere scroll horizontal):")
print(df_ancho) 
print('=' * 110)
# üí° Restablecer a True (comportamiento normal)
pd.set_option("display.expand_frame_repr", True)
print('DataFrame ancho en m√∫ltiples lineas')
print(df_ancho)
````
