# Constructores alternativos de DataFrames üë∑
Ademas del constructor est√°ndar ```pd.DataFrame()```, pandas ofrece __m√©todos espec√≠ficos__ para crear Dataframes seg√∫n casos comunes.

## 1. M√©todo: ```from_dict()```
Creaci√≥n desde diccionarios, se puede tener una Orientaci√≥n controladaüß≠.

````python
import numpy as np
import pandas as pd
````

### 1.1. Orientaci√≥n por columnas (forma predetermianda o por defult)

````python

# Cada lista dentro de diccionario conforma una columna
datos = {
    "Equipo A" : [10, 20, 30], 
    "Equipo B" : [40, 50, 60] 
}

#Creaci√≥n est√°ndar
df_cols = pd.DataFrame.from_dict(datos)
print(f'Orientaci√≥n por columnas (las claves son columnas):\n{df_cols}')
````

### 1.2. Orientaci√≥n por √≠ndice
Los datos se __trasponen__, es √∫til renombrar las columnas manualmente.

````python
# Cada lista dentro del diccionario conforma una fila
df_rows = pd.DataFrame.from_dict(datos, orient= "index", columns = ["Col_1", "Col_2", "Col_3"])
print(f'Orientaci√≥n por columnas (las claves son filas):\n{df_rows}')
````

## 2.M√©todo: ```from_records()```
Desde registros estructurados. Este m√©todo est√° dise√±ado para __listas de tuplas__ o __arrays estructurados de NumPy__. Este permite especificar directamente qu√© campo (campo del array o columna) se convertir√° en el ___√≠ndice__ del DataFrame usando el par√°metros __index__. 

````python

#Convierte y establece el √≠ndice en un paso
data_records = np.array(
    [(1, 2.5, "Laptop"), (2, 3.0, "Mouse")],
    dtype=[("ID", "i4"), ("Peso", "f4"), ("Producto", "U10")]
)
print(f'Datos crudos (Array estructurado de NumPy):\n{data_records}')

print("=" * 80)
df_records = pd.DataFrame.from_records(data_records, index = "Producto")
print(f'üöÄDataFrame desde registros (Campo "producto" es en √≠ndice o index):\n{df_records}')
````
