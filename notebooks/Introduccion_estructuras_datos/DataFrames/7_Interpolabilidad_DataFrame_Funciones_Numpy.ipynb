# Intepolabilidad de DataFrame con Funci贸n NumPy
__Pandas__ est谩 integrada con __Numpy__ en su mayor铆a, dado que la gran parte de funciones NumPy pueden llamarse directamente sobre objetos ```Series``` o ```DataFrame```.

````python
import pandas as pd
import numpy as np
````

````python
# Preparaci贸n de DataFrame de ejemplo
np.random.seed(42)
df_ejemplo = pd.DataFrame(np.random.randn(5, 4), columns=["A", "B", "C", "D"])
print(" DataFrame Original (5x4):")
print(df_ejemplo)
````

# 1. Aplicaci贸n de funciones universales (ufuncs) М
Las funciones universales de NumPy (ufuncs) se aplican elemento por elementos a los ```DataFrames``` y ```Series```, manteniendo sus estructuras intactas.

### 1.1. Aplicamos la funci贸n exponenciad de Numpy ```np.exp()```

````python
df_ejemplo_exp = np.exp(df_ejemplo)
print(f'Resultado:\n{df_ejemplo_exp}')
````

### 1.2. Conversi贸n a Array de NumPy П

````python
array_puro = np.asarray(df_ejemplo)
print(f'Tipo de objeto resultante:\n{ type(array_puro) }')
print('=' * 50)
print(f'Array puro, sin etiquetas:\n{ array_puro }')
````

## 2. Alineaci贸n en funciones con multiples Series З
cuando se pasa __m谩s de una Serie__ a un funci贸n ```ufunc``` de NumPy, Pandas realiza la alineaci贸n autom谩tica por etiqueta antes de ejecutar la operaci贸n.

````python
#A. Alineaci贸n por etiquetas desordenadas
#  Series con el mismo 铆ndice, pero desordenado
ser1 = pd.Series([34, 20, 10], index=["a", "b", "c"])
ser2 = pd.Series([3, 5, 2], index=["b", "a", "c"]) # 'a' y 'b' est谩n cambiados

print(f' Series 1:\n{ser1}' )
print('=' * 35)
print(f' Series 2 (Desordenada):\n{ser2}')
print('=' * 35)
# Se realiza un alineado por etiqueta
residuo_alineado = np.remainder(ser1, ser2)
print(f'Resultado de np.remainder():\n{ residuo_alineado }')
````

````python
#B. Manejo de Nan y uni贸n de 铆ndices
ser3 = pd.Series([2, 4, 6], index=["b", "c", "d"])
print(f'Serie 1:\n{ ser1 }')
print('=' * 40)
print(f'Serie 2:\n{ser2}')
print('=' * 40)
residuos_nan = np.remainder(ser1, ser3)
print(f'Resultado de np.remainder(ser1, ser3):\n{ residuos_nan }')
````

## 3. Prioridad: Serie vs Index 锔

````python
# No hay alineaci贸n por etiquetas , los compara por posiicones
s_valores = pd.Series([1, 2, 3])
idx_valores = pd.Index([4, 5, 0])
print(f'Serie:\n{s_valores}')
print('=' * 60)
print(f'Index:\n{idx_valores}')
print('=' * 60)

#np.maximum compara elemento a elemento y devuelve el mayor entre de los dos
maximos = np.maximum(s_valores, idx_valores)
print(f'Resultado de np.maximums(s_valores, idx_valores):\n{maximos}')
print('=' * 60)
print( type(maximos) )
````
