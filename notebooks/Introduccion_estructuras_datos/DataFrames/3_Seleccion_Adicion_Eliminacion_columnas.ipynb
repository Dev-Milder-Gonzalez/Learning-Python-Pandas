# Selecci√≥n, adici√≥n y Eliminaci√≥n de ___columnas___ üñäÔ∏è 

````python
import numpy as np
import pandas as pd
````

# 1. Creaci√≥n y selecci√≥n de columnas 
 - __Creaci√≥n:__ En base a operaciones matem√°ticas
 - __Selecci√≥n:__ Operaciones l√≥gicas de otras columnas existentes.

````python
diccionario = {
    "Uno": pd.Series([1.0, 2.0, 3.0], index = ["A", "B", "C"]),
    "Dos": pd.Series([5.0, 5.0, 6.0, 7.0], index = ["A", "B", "C", "D"])
}

df = pd.DataFrame(diccionario)
print(f'‚ú®DataFrame original:\n{df}')
````

````python
#1Ô∏è‚É£Selecci√≥n: Obtener una columna
columna_uno = df["Uno"]
print(f'üîéPrimera columna del DF:\n{columna_uno}')
````

````python

#2Ô∏è‚É£Adicic√≥n 1: Nueva colunma calculada: operaci√≥n vectorizada
df["Tres"] = df["Uno"] * df["Dos"]
print(f'Operaci√≥n vectorizada, adici√≥n de columnas "Uno" y "Dos": \n{df}')
````

````python

#3Ô∏è‚É£Adici√≥n 2: Nueva columna l√≥gica (booleana)
df["flag"] = df["Uno"] > 2
print(f'Operaci√≥n logica con lo elementos de la columna "Uno":\n{df}')
````

## 2. Eliminaci√≥n de columnas

````python
#1. Opci√≥n A: Usando "del" (borrar sin devolver nada)
del df["Dos"]
print(f'DataFrame despu√©s de eliminar la columna "Dos": \n{df}')
````

````python
#2. Opci√≥n B: Usando "pop" (borrar y capturar la columna)
columna_Tres = df.pop("Tres")
print(f'DataFrame despu√©s de eliminar la columna "Tres": \n{df}')
````

## 3. Inserci√≥n

### 3.1. Inserci√≥n de valores escalares (n√∫mero o cadena de texto)

````python
df["foo"] = "bar"
print(f'DataFrame despu√©s de agregar un valor escalar que se propaga en toda la columna:\n{df}') 
````

### 3.2 Inserci√≥n de series
En el caso de que insertemos ```Series``` que no tiene el mismo √≠ndice que el DataFrame, pandas la alinear√° aumtom√°ticamente.

````python
#Seleccionamos los dos primeros elementos de la columna "Uno" del DF
serie = df["Uno"][:2]
#Luego creamos una nueva columna con los valores seleccionados de la columna "Uno"
df["uno_trunc"] = serie
print(f'üß©DataFrame con columna alineada:\n{df}')
````

### 3.3. Inserci√≥n en una posici√≥n espec√≠fica
Por defecto, las columnas se agregan al final (derecha). Pero podemos insertarlas es una posici√≥n espec√≠fica.

 - Podemos observar que las posiciones de las columnas van est√°n dentro del rango ```[0, cantidad_columnas -1]```. 
 - Para las filas es lo mismo estan dentro den rango ```[0, cantidad_filas -1]```.

````python
df.insert(1, "flag_copia", df["flag"])
print(f'üéàDataFrame con una columna insertada en la posici√≥n 1:\n{df}')
````
